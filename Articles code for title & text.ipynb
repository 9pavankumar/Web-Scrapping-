```python
# Import the 'os' module for operating system-related functions
import os

# Import the 'requests' library for making HTTP requests
import requests

# Import 'BeautifulSoup' from the 'bs4' library for web scraping
from bs4 import BeautifulSoup

# Import the 'pandas' library for data manipulation and analysis
import pandas as pd

```

```python

# Output folder path
output_folder = r"C:\Users\1pava\Documents\ALL PROJECTS\Web Scrapping\final_articles"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Load the input DataFrame from Excel
input_df = pd.read_excel(r"C:\Users\1pava\Documents\ALL PROJECTS\Web Scrapping\Output Data Structure.xlsx")
```

```python

# Iterate through each row in the input DataFrame
for index, row in input_df.iterrows():
    url = row['URL']

    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Extract the title
        title_element = soup.find('title')
        article_title = title_element.text if title_element else "No title found"

        # Extract the main content of the article (adjust the class or tag accordingly)
        main_content_element = soup.find('div', class_='td-post-content')  # Adjust the class accordingly
        if main_content_element:
            # Remove unwanted elements (e.g., headers, footers)
            unwanted_elements = main_content_element.find_all(['header', 'footer'])  # Add more if needed
            for unwanted_element in unwanted_elements:
                unwanted_element.decompose()

            # Extract the cleaned article content
            article_content = main_content_element.get_text().strip()

            # Create a text file with the title as its file name in the specified output folder
            file_name = os.path.join(output_folder, f"{row['URL_ID']}.txt")
            with open(file_name, 'w', encoding='utf-8') as file:
                file.write(f"Title: {article_title}\n\n")
                file.write(f"Text: {article_content}\n")

            print(f"File '{file_name}' created successfully.")
        else:
            print(f"No main content found for URL_ID {row['URL_ID']}.")
    else:
        print(f"Failed to fetch URL {url}. Status code: {response.status_code}")
```
